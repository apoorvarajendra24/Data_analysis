```python
#  Data Visualization with Seaborn library

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

```python
# line plot
sns.lineplot(x='size',y='tip', data=tips)
plt.show()
```

```python
# Categorical plot
# Bar plot
plt.subplot(1,2,1)
sns.barplot(x='sex',y='tip',color='teal',data=tips)
plt.title("Bar plot for tips given by male and female")

plt.subplot(1,2,2)
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar plot for total bill by day")
plt.tight_layout()
plt.show()
```

```python
# Box plot
sns.boxplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
# Volin plot 
sns.violinplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
# Histogram plot

sns.histplot(tips['total_bill'],bins=10, kde=True)
plt.title('Histogram of Total Bill by Days')
plt.show()
```

```python
# Kde plot

import numpy as np
data = np.random.randn(1000)
sns.kdeplot(data, fill=True, color='blue')
plt.title("KDE Plot for random numbers")
plt.show()
```

```python
# Pair plot 
sns.pairplot(tips)
```

```python
# Heatmap

corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
