```python
# Matplotlib library
import matplotlib.pyplot as plt
# line plot
x = [1,2,3,4,5,6]
y = [12,10,7,6,4,1]
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.show()
```

```python
# A customized line plot
x = [1,2,4,5]
y = [2,4,6,8]

plt.plot(x,y,color='green',linestyle='-.', marker='o', linewidth=2, markersize=5)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Customized Line Plot")
plt.grid(True)           # creates a grid
plt.show()
```

```python
# plotting both lines in same plot
import matplotlib.pyplot as plt

# Data points
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]  # First line
y2 = [5, 15, 25, 35, 45]   # Second line

# Plot both lines
plt.plot(x, y1, label='Line 1', color='blue', marker='o')
plt.plot(x, y2, label='Line 2', color='red', linestyle='dashed', marker='s')

# Labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plots')

# Show legend
plt.legend()             # used to display a legend (label for plotted lines, bars, etc.) in a Matplotlib graph.

# Display the plot
plt.show()
```

```python
# Creating multiple subplots
import matplotlib.pyplot as plt

# First subplot (1 row, 2 columns, first plot)
plt.subplot(1, 2, 1)  
plt.plot([1, 2, 3], [4, 5, 6], color='blue')
plt.title("First Plot")
plt.xlabel("X- axis")
plt.ylabel("Y- axis")
# Second subplot (1 row, 2 columns, second plot)
plt.subplot(1, 2, 2)  
plt.plot([1, 2, 3], [6, 5, 4], color='green',linestyle='-.')
plt.title("Second Plot")
plt.xlabel("X- axis")
plt.ylabel("Y- axis")
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

```python
# Bar plot
categories = ['A','B','C','D']
values = [5,4,2,8]

plt.figure()
plt.subplot(1,2,1)
plt.bar(categories,values, color='purple', width=0.5)          # a vertical bar chart
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example 1')

plt.subplot(1,2,2)
plt.barh(categories,width=values, color='purple')          # a horizontal bar chart
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example 2')
plt.tight_layout()
plt.show()
```

```python
# grouped bar plot
import numpy as np
# Data
categories = ['A', 'B', 'C']
values1 = [10, 20, 15]
values2 = [12, 18, 22]

# X positions
x = np.arange(len(categories))

# Create bars
plt.bar(x - 0.2, values1, width=0.4, label="Group 1", color='blue')
plt.bar(x + 0.2, values2, width=0.4, label="Group 2", color='red')

# Labels
plt.xticks(x, categories)  # Set category names on x-axis
plt.legend()
plt.title("Grouped Bar Plot")
plt.show()
```

```python
# Histogram plot

import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.randn(1000)  # 1000 values from a normal distribution

# Create histogram
plt.hist(data, bins=30, color='blue', edgecolor='black')

# Labels and title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")

# Show plot
plt.show()
```

```python
# histogram with seaborn density curv
import seaborn as sns
sns.histplot(data, bins=30, kde=True, color='purple')  # kde=True adds a smooth density curve
plt.show()
```

```python
# overlapping histogram
data1 = np.random.randn(1000) * 0.5  # Dataset 1
data2 = np.random.randn(1000) * 1.5  # Dataset 2

plt.hist(data1, bins=30, alpha=0.5, label="Data 1", color='blue')
plt.hist(data2, bins=30, alpha=0.5, label="Data 2", color='red')

plt.legend()
plt.title("Overlapping Histograms")
plt.show()
```

```python
# horizontal histogram
plt.hist(data, bins=30, orientation='horizontal', color='orange')
plt.show()
```

```python
# Multiple scatter plot in single graph

x1 = np.random.rand(30) * 30
y1 = np.random.rand(30) * 50

x2 = np.random.rand(30) * 30
y2 = np.random.rand(30) * 50

plt.scatter(x1,y1,color='blue',edgecolors='black',label='Dataset-1')
plt.scatter(x2,y2,color='green',edgecolors='black',label='Dataset-2')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Multiple Scatter Plot")
plt.legend(loc='upper right')
plt.show()
```

```python
# Scatter plot 
import numpy as np

# Random data
x = np.random.rand(50) * 10  # 50 random values for x (0-10)
y = np.random.rand(50) * 100  # 50 random values for y (0-100)
sizes = np.random.rand(50) * 200  # Size of points
colors = np.random.rand(50)  # Colors of points

plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=1, edgecolors='black')

plt.colorbar(label="Color Scale")  # Add color legend
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Customized Scatter Plot")
plt.show()

"""
✔ s=sizes → Varying point sizes.
✔ c=colors, cmap='viridis' → Color gradient.
✔ alpha=0.7 → Transparency for better visibility.
"""
```

```python
# Pie chart - A pie chart is used to show the proportion of categories in a dataset.

import matplotlib.pyplot as plt

# Data
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [25, 35, 20, 20]
# Create pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
# Title and show
plt.title("Basic Pie Chart")
plt.show()
```

```python
colors = ['red', 'yellow', 'pink', 'brown']
explode = (0, 0.1, 0, 0)  # "Banana" slice pulled out

plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, explode=explode, shadow=True)
plt.title("Customized Pie Chart")
plt.show()
```
