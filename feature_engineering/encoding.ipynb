One Hot Encoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# creating instance of Onehotencoder
encoder = OneHotEncoder()

# Creating dataframe
df_color = pd.DataFrame(
    {'color': ['red', 'blue', 'blue', 'green','red','purple']})

# Perform fit and transform
encoded = encoder.fit_transform(df_color[['color']]).toarray()   # albhabetical sorting
encoded
```

```python
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
df_encoded = pd.concat([df_color,encoded_df],axis=1)
df_encoded
```

```python
# adding a new feature to the data frame
new_addition = encoder.transform([['purple']]).toarray()

new_addition_df = pd.DataFrame(new_addition, columns=encoder.get_feature_names_out(['color']))
df_new_row = pd.DataFrame({'color':['purple']})
new_color_df = pd.concat([new_addition_df,df_new_row],axis=1)

new_row_df = pd.concat([df_encoded,new_color_df],axis=1)
new_row_df
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

lbl_encoder = LabelEncoder()
encoded = lbl_encoder.fit_transform(df_color[['color']])
df_encoded = pd.DataFrame({'color': df_color['color'], 'encoded': encoded})
df_encoded
```

Ordinal encoding

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

# Sample DataFrame
df = pd.DataFrame({'Quality': ['Low', 'Medium', 'High', 'Medium', 'Low', 'High']})

# Define category order
categories = [['Low', 'Medium', 'High']]  # Ensures order is maintained

# Initialize and fit OrdinalEncoder
encoder = OrdinalEncoder(categories=categories)
df['Quality_encoded'] = encoder.fit_transform(df[['Quality']])
df
```

Target Guided Encoding

```python
df = pd.DataFrame({'City': ['A', 'B', 'C', 'A', 'D', 'C', 'B', 'A', 'C','D'],
        'Price': [100, 200, 150, 120, 180, 210, 130, 170, 180, 197]})  # Target variable

mean_encoded = df.groupby('City')['Price'].mean().to_dict()
df['City_encoded'] = df['City'].map(mean_encoded)
df
```

```python
df_encoded = df[['Price','City_encoded']]
df_encoded
```
